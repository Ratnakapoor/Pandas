```python
# Calculate the average math score & Reading score
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

Average_math = renamed_df["math_score"].mean()
Average_reading = renamed_df["reading_score"].mean()
Average_overall = (( Average_math + Average_reading ) / 2)
```

```python
# Calculate the percentage of students with a passing math score (70 or greater) & Reading score 70 and greater

is_70_math =  renamed_df["math_score"] > 69
new_df= renamed_df[is_70_math]
math_cnt = new_df["math_score"].count()
Passing_math = (( math_cnt / total_students ) * 100)

is_70_read =  renamed_df["reading_score"] > 69
new_df= renamed_df[is_70_read]
read_cnt = new_df["reading_score"].count()
Passing_read = (( read_cnt / total_students ) * 100)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Create a  Dataset for Top Performing school wth filters for pass in Math & Reading

school_data_complete = student_data[['school_name', 'Student ID', 'reading_score', 'math_score']].merge(school_data[['school_name', 'type', 'budget']], on='school_name', how='left')
school_data_complete['Mpass'] = np.where(school_data_complete['math_score']>=70, 'yes', 'no')
school_data_complete['Rpass'] = np.where(school_data_complete['reading_score']>=70, 'yes', 'no')
school_data_complete.head(n=10)
```

```python
#School Data Frame Group by School name 

school_df = school_data_complete.groupby(["school_name"])

# 2nd data set for Math pass count 
df = school_data_complete[school_data_complete['Mpass'] == "yes"]
math_df = df.groupby(["school_name"])

# 3rd data set for Reading pass count 
df = school_data_complete[school_data_complete['Rpass'] == "yes"]
read_df = df.groupby(["school_name"])
```

```python
# Aggregate the nos to create a Summary 

per_school_summary = pd.DataFrame({'School Type' :        school_df.type.unique(),
                   'Total Student':       school_df["Student ID"].count() ,
                   'Total School Budget': school_df.budget.unique(), 
                   'Per Student Budget' : abs(school_df["budget"].min() / school_df["Student ID"].count()) ,
                   'Avg Math Score' :     round(school_df["math_score"].mean(),2),
                   'Avg Read Score' :     round(school_df["reading_score"].mean(),2),
                                   
                   '% Passing Math' :     round((( math_df["Student ID"].count() / school_df["Student ID"].count() ) * 100) , 2),
                                   
                   '% Passing Reading ':  round((( read_df["Student ID"].count() / school_df["Student ID"].count() ) * 100) , 2),
                                   
                   '% Overall Pass' :   (((( math_df["Student ID"].count() / school_df["Student ID"].count() ) * 100)  +
                                        (( read_df["Student ID"].count() / school_df["Student ID"].count() ) * 100)) / 2 ) 
                  }) 
                                                
per_school_summary
```

```python
#Sort and display the top five schools in overall passing rate

top_schools = per_school_summary.sort_values(["% Overall Pass"], ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools (By Passing Rate)

```python
#Sort and display the five worst schools in overall passing rate

bottom_schools = per_school_summary.sort_values(["% Overall Pass"], ascending=True)
bottom_schools.head(5)
```
